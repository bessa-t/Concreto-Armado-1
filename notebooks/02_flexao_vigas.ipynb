# Verificacao de flexao de vigas

Memorial executavel para verificacao preliminar de vigas retangulares em concreto armado, usando a ABNT NBR 6118:2023 Versao Corrigida 2:2024 como referencia normativa do projeto.

```python
from pathlib import Path
import sys
import pandas as pd

PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

from calculos.flexao import verificar_flexao_viga
from calculos.materiais import Aco, Concreto
from calculos.secoes import SecaoRetangular
from memorial.relatorio import montar_linha_flexao
```

## Dados de entrada

```python
concreto = Concreto(fck=30, E=30672)
aco = Aco(fyk=500)

vigas = pd.read_csv(PROJECT_ROOT / 'dados/vigas.csv')
vigas
```

## Processamento

```python
linhas = []

for viga in vigas.itertuples(index=False):
    resultado = verificar_flexao_viga(
        secao=SecaoRetangular(bw=viga.bw_cm, h=viga.h_cm),
        concreto=concreto,
        aco=aco,
        mk_kn_m=viga.mk_kn_m,
        cobrimento_cm=viga.cobrimento_cm,
        diametro_estribo_mm=viga.diametro_estribo_mm,
        diametro_barra_estimado_mm=viga.diametro_barra_estimado_mm,
    )
    linhas.append(montar_linha_flexao(viga.nome, resultado))

resumo = pd.DataFrame(linhas)
resumo
```
